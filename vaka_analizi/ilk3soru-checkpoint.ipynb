```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
```

```python
data = pd.read_csv("filtered.csv", index_col=0)
```

```python
# Diferansiyel İfade Analizi
def differential_expression(data):
    # Mock ve SARS-CoV sütunları
    mock_cols = [col for col in data.columns if col.startswith('mock_')]
    sars_cols = [col for col in data.columns if col.startswith('sars_')]
    
    # t-test yaparak p-value hesaplama
    p_values = data.apply(lambda row: ttest_ind(row[mock_cols], row[sars_cols])[1], axis=1)
    
    results = pd.DataFrame({
        'p_value': p_values,
        'significantly_changed': p_values < 0.05
    })
    
    return results
```

```python
de_results = differential_expression(data)
```

```python
significant_genes = de_results[de_results['significantly_changed']]
significant_genes
```

```python
significant_genes.to_csv("significant_genes.csv")
```

```python
converted_alias_df = significant_genes.index.to_frame(name='converted_alias')
converted_alias_df.to_csv("enrichment.csv", index=False)
```
